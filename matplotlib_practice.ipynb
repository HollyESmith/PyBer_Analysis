```python
# import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, color="blue", linewidth=2, marker="", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.xlim(0, 5)
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, color="green", linewidth=2, marker="D", label = 'Boston')
# Create labels for the x and y axes.
plt.xlabel("Date",fontname="Arial", size=12,fontweight="bold")
plt.ylabel("Fare($)",fontname="Arial", size=12,fontweight="bold")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month",fontname="Arial", size=12,fontweight="bold")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
plt.xlim(0, 11)
```

```python
#Create the plot
fig, ax = plt.subplots()

# x axis values
x =["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
  
# y axis values
y =[10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]

#set color
plt.plot(x, y, color='green', marker='D', linewidth=2)
     
# creating the canvas
fig = plt.figure()

#add a label that will be added to the legend
ax.plot(x, y, label='line', color='green')

#add a label to the x-axis, y-axis
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')

#add a legend
ax.legend("Boston")

#add y-axis limit
ax.set_ylim(0,45)

#add title
ax.set_title('PyBer Fare by Month')

#add grid lines
ax.grid()
```
